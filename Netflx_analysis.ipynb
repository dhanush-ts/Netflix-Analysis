## Netflix Analysis using Pandas.

```python
import pandas as pd
```

```python
df=pd.read_csv(r"C:\Pandas\Netflix_analysis\netflix_titles.csv")
```

### Task 1 : What is the most popular release year for movies on Netflix?

```python
df.head(10)
```

```python
df1=df[df.type=="Movie"]
df2=df1.groupby("release_year").count().reset_index().sort_values(by="type",ascending=0)[["release_year","type"]]
df2
```

```python
df2[df2.type==df2.type.max()]
```

### Task 2 : What year did Netflix add the most content to its platform?

```python
df["date_add"]=pd.to_datetime(df.date_added,format='mixed')
df["add_yr"]=df.date_add.dt.year
df
```

```python
df3=df.groupby("add_yr").count().reset_index().sort_values(by="type",ascending=0)[["add_yr","type"]]
df3
```

```python
df3[df3.type==df3.type.max()]
```

### Task 3: What is the most popular month to add new content?

```python
df["month"]=df.date_add.dt.month
df
```

```python
df4=df.groupby("month").count().reset_index().sort_values(by="type",ascending=0)[["month","type"]]
df4
```

```python
df4[df4.type==df4.type.max()]
```
